## The Jeans Mass

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as mp
from google.colab import drive
import os
%matplotlib inline
```

## Jeans Mass vs Temperature at fixed lenght (100Mpc)

```python
mass = (2*(10**9), 9.8*(10**9), 2*(10**10))
temp = (10**4, 5*(10**4), 10**5)
```

```python
fig, graf1 = mp.subplots()

graf1.plot(temp, mass, marker='o', linestyle='None')
graf1.set(xlabel='T (K)', ylabel='M (kg)', title='Jeans Mass vs. Temperature')

# Fit a linear regression line
z = np.polyfit(temp, mass, 1)
p = np.poly1d(z)
graf1.plot(temp, p(temp), "r--")  # Plot the tendency line in red dashed style

mp.show()

# Extract the slope (angular coefficient)
K = z[0]
print(f"K = {K} kg/K")
```

M = K T^a con a = 1


## Jeans Mass vs lenght at fixed Temperature (10^4 K)

M = K T^a (l/10kpc)^b

```python
massa = (1.05*(10**9), 1.6*(10**9), 2.3*(10**9), 2.95*(10**9), 4.9*(10**9))
length = (10, 15, 20, 25, 30)
```

```python
fig, graf2 = mp.subplots()

graf2.plot(length, massa, marker='o', linestyle='None')
graf2.set(xlabel='length (kpc)', ylabel='M (kg)', title='Jeans Mass vs. length')

# Fit a linear regression line
z = np.polyfit(length, massa, 1)
p = np.poly1d(z)
graf2.plot(length, p(length), "--", label='1st')  # Plot the tendency line in red dashed style

# Fit a 2nd grade plynomial regression line
z = np.polyfit(length, massa, 2)
p = np.poly1d(z)
graf2.plot(length, p(length), "--",label='2nd')

# Fit a 3rd grade plynomial regression line
z = np.polyfit(length, massa, 3)
p = np.poly1d(z)
graf2.plot(length, p(length), "--", label='3rd')

# Fit a 4th grade plynomial regression line
z = np.polyfit(length, massa, 4)
p = np.poly1d(z)
graf2.plot(length, p(length), "--", label='4th')

mp.legend()
mp.show()
```
