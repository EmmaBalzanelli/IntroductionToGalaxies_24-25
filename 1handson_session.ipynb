```python
from astropy.table import Table
sim = Table.read("/content/drive/My Drive/lab Astro MAGISTRALE/Hands-on Session 1/MR7_z0.0_L150_h0.7.fits")
print(sim)
Mhalo = np.log10(sim["MVIR"])
Vif = sim["INFALLVMAX"]
```

```python
print(Mhalo)
```

```python
from astropy.table import Table
tab = Table.read("/content/drive/My Drive/lab Astro MAGISTRALE/Hands-on Session 1/SMF_Baldry2012.txt", format="ascii.commented_header")
Mstar = tab["Mass"]
Ndens = tab["Density"]
print(tab)
totdens = np.sum(Ndens)
totgal = totdens*150**3
print(totdens)
print(totgal)
```

```python
mp.plot(Mstar, Ndens)
mp.xlabel("log(M)")
mp.ylabel("Density")
mp.title("Stellar Mass Function (SMF)")
```

```python
vol = 150**3
Ngalaxsies = vol*Ndens*0.2
print(Ngalaxsies)
```

```python
totgalaxsies = np.sum(Ngalaxsies)
print(totgalaxsies)
```

```python
sorted_Mhalo = []
for i in range(len(Mhalo)):
  sorted_Mhalo.append(Mhalo[i])
sorted_Mhalo.sort()
print(sorted_Mhalo)
```

```python
gal_mass = []
gal_number = []
for i in range(len(Mstar)):
  gal_mass.append(Mstar[i])
  gal_number.append(Ngalaxsies[i])
print(gal_mass)
print(gal_number)
```

```python
fig, graf = mp.subplots()
graf.bar(gal_mass, gal_number, 0.2)
mp.show()
```

```python
import random
mass_vector = np.array(())

for i in range(len(gal_mass)):
  temp_mass_vector = np.random.uniform(gal_mass[i] - 0.1, gal_mass[i] + 0.1, size = int(gal_number[i]))
  mass_vector = np.concatenate((mass_vector, temp_mass_vector))

print(mass_vector)
print(len(mass_vector))
```

```python
sorted_mass_vector = np.sort(mass_vector)
print(sorted_mass_vector)
```

```python
less = len(sorted_mass_vector) - len(sorted_Mhalo)

sorted_mass_vector_cut = sorted_mass_vector[less:]
print(sorted_mass_vector_cut)
print(len(sorted_mass_vector_cut))
print(len(sorted_Mhalo))
```

```python
mp.plot(sorted_Mhalo, sorted_mass_vector_cut)
mp.xlabel("log(Mhalo)")
mp.ylabel("log(Mstar)")
mp.title("Mstar and Mhalo relation")
```

```python
fig, grafico = mp.subplots()
grafico.plot(sorted_Mhalo, sorted_mass_vector_cut-sorted_Mhalo)
grafico.set(xlabel='log(Mhalo)', ylabel='log(Mstar/Mhalo)', title='Mstar and Mhalo ratio')
```
